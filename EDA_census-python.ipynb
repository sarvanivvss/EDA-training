```python
literacy_column_name = census.columns[11]
population_column_name = census.columns[2]
state_column_name=census.columns[3]
# Calculate literacy rate
# Calculate literacy rate as a percentage
row_index = 2  # 6th row (Python uses 0-based indexing)
selected_row = census.iloc[row_index]

# Calculate literacy rate as a percentage using values from the selected row
literacy_rate = (selected_row[literacy_column_name] / selected_row[population_column_name]) * 100

# Find the state with the lowest literacy rate
lowest_literacy_state = census.loc[census['Literacy Rate'].idxmin(), state_column_name]


# Visualize the literacy rates across all states
plt.figure(figsize=(12, 6))
sns.barplot(x=state_column_name, y='Literacy Rate', data=census)
plt.title('Literacy Rates Across States')
plt.xlabel('State')
plt.ylabel('Literacy Rate')
plt.xticks(rotation=90)
plt.show()
```
